## Run `train_baseline.py`

```python
from datetime import datetime
from pathlib import Path
import subprocess
import sys

repo_root = Path.cwd().resolve()
if not (repo_root / 'train_baseline.py').exists():
    candidate = repo_root.parent
    if (candidate / 'train_baseline.py').exists():
        repo_root = candidate

if not (repo_root / 'train_baseline.py').exists():
    raise FileNotFoundError(f'Could not find train_baseline.py from {Path.cwd()}')

log_dir = repo_root / 'runs' / 'train_logs'
log_dir.mkdir(parents=True, exist_ok=True)
log_path = log_dir / f"train_baseline_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"
print(f"Repo root: {repo_root}")
print(f"Logging to: {log_path}")

cmd = [sys.executable, str(repo_root / 'train_baseline.py')]
with log_path.open('w', encoding='utf-8') as f:
    process = subprocess.Popen(
        cmd,
        cwd=str(repo_root),
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
    )
    for line in process.stdout:
        print(line, end='')
        f.write(line)
    return_code = process.wait()

print(f"\nFinished with return code: {return_code}")
if return_code != 0:
    raise RuntimeError(f"train_baseline.py failed with return code {return_code}")

```
